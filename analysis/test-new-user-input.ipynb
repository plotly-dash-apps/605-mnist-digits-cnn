```python
import pickle
import pandas as pd
import numpy as np
from keras.models import load_model
from numpy import argmax
```

```python
## read in our pickle file
filename = open('inputs/user-input-digit.pkl', 'rb')
array_to_data_output = pickle.load(filename)
filename.close()
```

```python
# what is the digit type
print(type(array_to_data_output))
print(array_to_data_output.shape)
```

```python
# convert the user input to the format expected by the model
some_digit_array = np.reshape(array_to_data_output.values, -1)
print(some_digit_array[:150])
```

```python
# load and prepare the image
def load_image(img):
	# load the image
	# img = load_img(filename, grayscale=True, target_size=(28, 28))
    
	# convert to array
	# img = img_to_array(img)
    
	# reshape into a single sample with 1 channel
	img = img.reshape(1, 28, 28, 1)
	# prepare pixel data
	img = img.astype('float32')
	img = img / 255.0
	return img
```

```python
# load an image and predict the class
def run_example(filename):
	# load the image
	img = load_image(filename)
	# load model
	model = load_model('models/cnn_model_4.h5')
	# predict the class
	predict_value = model.predict(img)
	digit = argmax(predict_value)
	print(digit)
```

```python
# entry point, run the example
run_example(some_digit_array)
```
